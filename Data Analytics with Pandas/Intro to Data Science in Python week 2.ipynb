```python
def chickenpox_by_sex():
    '''1- You divide the dataframe into two: one for male, the other for female. 
          On each dataframe apply the correct vaccination filtering (>=1).

       2- On each dataframe in part 1, calculate the ratio of children who contracted 
          chickenpox vs those who did not. Notice that the denominator is not equal to 
          the number of total children in each dataframe.

       3- Return the values you calculated with keys "male" and "female" respectively in a dictionary.'''
    
    # YOUR CODE HERE
    # m_vaxxed, m_unvaxxed, m_vaxxed_wpox, m_vaxxed_wopox
    #f_vaxxed, f_unvaxxed, f_vaxxed_wpox, f_vaxxed_wopox
    df_cpox = pd.read_csv(jab_file, usecols=['SEQNUMC', 'HAD_CPOX', 'P_NUMVRC', 'SEX'])
    male_vax_stats = {'m_vaxxed': 0, 'm_vaxxed_wpox': 0, 'm_vaxxed_wopox': 0,
                      'm_unvaxxed': 0, 'm_unvaxxed_wpox': 0, 'm_unvaxxed_wopox': 0}
    female_vax_stats = {'f_vaxxed': 0, 'f_vaxxed_wpox': 0, 'f_vaxxed_wopox': 0,
                        'f_unvaxxed': 0, 'f_unvaxxed_wpox': 0, 'f_unvaxxed_wopox': 0}
    
    pox_rates_by_sex ={}
    
    df_cpox_t = df_cpox['P_NUMVRC'] >= 0
    df_cpox_tot = df_cpox[df_cpox_t]
    
    df_cpox_m = df_cpox['SEX'] == 1  # create the boolean mask
    df_cpox_males = df_cpox[df_cpox_m]  # apply the boolean mask
    df_cpox_vm = df_cpox_males['P_NUMVRC'] >= 1 # create additional boolean mask
    df_cpox_vmales = df_cpox_males[df_cpox_vm]  # apply the boolean mask
    df_cpox_vmyes_mask = df_cpox_vmales['HAD_CPOX'] == 1
    df_cpox_vmyes = df_cpox_vmales[df_cpox_vmyes_mask]
    df_cpox_vmno_mask = df_cpox_vmales['HAD_CPOX'] == 2
    df_cpox_vmno = df_cpox_vmales[df_cpox_vmno_mask]
    
    
    df_cpox_uvm = df_cpox_males['P_NUMVRC'] == 0
    df_cpox_uvmales = df_cpox_males[df_cpox_uvm]
    df_cpox_uvmyes_mask = df_cpox_uvmales['HAD_CPOX'] == 1
    df_cpox_uvmyes = df_cpox_uvmales[df_cpox_uvmyes_mask]
    
    male_vax_stats['m_vaxxed'] = len(df_cpox_vmales)
    male_vax_stats['m_vaxxed_wpox'] = len(df_cpox_vmyes)
    male_vax_stats['m_vaxxed_wopox'] = len(df_cpox_vmno)
    male_vax_stats['m_unvaxxed'] = len(df_cpox_uvmales)
    male_vax_stats['m_unvaxxed_wpox'] = len(df_cpox_uvmyes)
    male_vax_stats['m_unvaxxed_wopox'] = male_vax_stats['m_unvaxxed'] - male_vax_stats['m_unvaxxed_wpox']
    
    df_cpox_f = df_cpox['SEX'] == 2  # create the boolean mask
    df_cpox_females = df_cpox[df_cpox_f]  # apply the boolean mask
    df_cpox_vfm = df_cpox_females['P_NUMVRC'] >= 1 # create additional boolean mask
    df_cpox_vfemales = df_cpox_females[df_cpox_vfm]  # apply the boolean mask
    df_cpox_vfmyes_mask = df_cpox_vfemales['HAD_CPOX'] == 1
    df_cpox_vfmyes = df_cpox_vfemales[df_cpox_vfmyes_mask]
    df_cpox_vfno_mask = df_cpox_vfemales['HAD_CPOX'] == 2
    df_cpox_vfno = df_cpox_vfemales[df_cpox_vfno_mask]
    
    df_cpox_ufvm = df_cpox_females['P_NUMVRC'] == 0
    df_cpox_uvfemales = df_cpox_females[df_cpox_ufvm]
    df_cpox_uvfmyes_mask = df_cpox_uvfemales['HAD_CPOX'] == 1
    df_cpox_uvfmyes = df_cpox_uvfemales[df_cpox_uvfmyes_mask]
    
    female_vax_stats['f_vaxxed'] = len(df_cpox_vfemales)
    female_vax_stats['f_vaxxed_wpox'] = len(df_cpox_vfmyes)
    female_vax_stats['f_vaxxed_wopox'] = len(df_cpox_vfno)
    female_vax_stats['f_unvaxxed'] = len(df_cpox_uvfemales)
    female_vax_stats['f_unvaxxed_wpox'] = len(df_cpox_uvfmyes)
    female_vax_stats['f_unvaxxed_wopox'] = female_vax_stats['f_unvaxxed'] - female_vax_stats['f_unvaxxed_wpox']

    # we need the vaxed children
    # ratio of those who contracted pox versue those who didn't by sex
    male_ratio = male_vax_stats['m_vaxxed_wpox'] / male_vax_stats['m_vaxxed_wopox']
    female_ratio = female_vax_stats['f_vaxxed_wpox'] / female_vax_stats['f_vaxxed_wopox']
    
    pox_rates_by_sex["male"] = male_ratio
    pox_rates_by_sex["female"] = female_ratio
    
    # {'male': 0.0096xxxxxxxxxx2664, 'female': 0.0077xxxxxxxxxx9565}
    # correct female 'female': 0.0077918259335489565
    print(df_cpox_vfno.head())
    print(len(df_cpox_vfno))
    print(len(df_cpox_tot))
    print(male_vax_stats['m_vaxxed'] + female_vax_stats['f_vaxxed'] + male_vax_stats['m_unvaxxed'] + female_vax_stats['f_unvaxxed'])
    print(male_ratio)
    print(female_ratio)
    print(pox_rates_by_sex)
    #print(df_cpox_vmyes_mask.head(10))
    #print(df_cpox_vmyes_mask.head(10))
    print(male_vax_stats)
    print(female_vax_stats)
    #return pox_rates_by_sex
    #raise NotImplementedError()
```
