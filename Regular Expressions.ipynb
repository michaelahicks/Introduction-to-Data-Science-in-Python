# Part A

Find a list of all of the names in the following string using regex.

```python
import re
def names():
    simple_string = """Amy is 5 years old, and her sister Mary is 2 years old. 
    Ruth and Peter, their parents, have 3 kids."""

    # YOUR CODE HERE
    new_lst = simple_string.split(" ")
    
    names = []
    for word in new_lst:
        if re.search('[A-Z]\w*', word):
            names.append(word)
    return names
    #re.search('\')
    raise NotImplementedError()
```

# Part B

The dataset file in assets/grades.txt contains a line separated list of people with their grade in a class. Create a regex to generate a list of just those students who received a B in the course.

```python
import re
grade_file = '/home/bigboss/Documents/Python/Python for Data Analysis/assignments/assignment1/assets/grades.txt'

'''
file = open(grade_file, "r")
grades = file.read()
print(grades)
print("*"*10)
#print(type(grades))
file.close()
'''

def grades():
    with open(grade_file, "r") as file:
        grades = file.read()
        grades = grades.split("\n")
        name_lst = []
        for stu in grades:
            if re.search(':\s[B]', stu):
                stu = re.search('\w*\s\w*', stu).group()
                name_lst.append(stu)
        print(name_lst)
        print("*"*10)
        print(len(name_lst))
    return name_lst

    # YOUR CODE HERE
    raise NotImplementedError()
```
